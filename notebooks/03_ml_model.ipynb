```python
import pandas as pd

from sklearn.model_selection import train_test_split, cross_val_score
#from sklearn.ensemble import RandomForestClassifier

from lightgbm import LGBMClassifier

from sklearn.metrics import accuracy_score, classification_report

from sklearn.preprocessing import StandardScaler
```

```python
ruta = r'D:\0_Respaldo\0_Proyectos_2024\ML_proyects\titanic\titanic\titanic\data\clean\ready_2.parquet'
```

```python
data = pd.read_parquet(ruta)
```

```python
data.info()
data.describe()
```

```python
df = data.copy()
```

```python
df.head(3)
```

```python
x = df.drop(['Survived'], axis=1) #Features
y = df['Survived'] #target

X_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 28)
```

```python
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    'n_estimators': randint(100, 1500),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 10),
    'num_leaves': randint(20, 100),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'min_child_samples': randint(20, 100),
    'reg_alpha': uniform(0, 1),
    'reg_lambda': uniform(0, 1)
}
```

```python
lgbm = LGBMClassifier(random_state=42)
```

```python
model = RandomizedSearchCV(
    estimator=lgbm,
    param_distributions=param_dist,
    n_iter=1000,  # número de combinaciones de parámetros a probar
    cv=5,  # número de pliegues para la validación cruzada
    verbose=2,
    random_state=42,
    n_jobs=-1  # usar todos los núcleos disponibles
)
```

```python
model.fit(X_train, y_train)
```

```python
print("Mejores parámetros encontrados:")
print(model.best_params_)
print("\nMejor puntaje:")
print(model.best_score_)
```

```python
params = {
    'n_estimators' : 145,
    'learning_rate': 0.073,
    'max_depth': 3,
    'num_leaves' : 74,
    'subsample' : 0.8,
    'colsample_bytree': 0.85,
    'min_child_samples': 22,
    'reg_alpha': 0.176,
    'reg_lambda': 0.23
}
```

```python
model = LGBMClassifier(**params)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x_test)

# Evaluar el rendimiento
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Imprimir un informe de clasificación
print(classification_report(y_test, y_pred))
```
